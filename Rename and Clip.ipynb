```python
import os
import glob
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
```

```python
def rename_files(path='/Users/swapnilsinghavi/CMU_Course' \
                 'Work/Spring2018/LSMA/Project/all_videos/'):
    '''
    path = path where the files exist
    '''
    files = os.listdir(path)
    for file in files:
        target_name = file
        target_name = target_name.replace("person","p")
        target_name = target_name.replace("boxing","b")
        target_name = target_name.replace("running","r")
        target_name = target_name.replace("handwaving","hw")
        target_name = target_name.replace("walking","w")
        target_name = target_name.replace("handclapping","hc")
        target_name = target_name.replace("jogging","j")
        target_name = target_name.replace("uncomp","uc")
           
        os.rename(os.path.join(path,file),os.path.join(path,target_name))
#        print(file, target_name)
```

```python
rename_files()
```

```python
def clip_files(path='/Users/swapnilsinghavi/CMU_Course' \
                 'Work/Spring2018/LSMA/Project/all_videos/',
              target_path='/Users/swapnilsinghavi/CMU_Course' \
                 'Work/Spring2018/LSMA/Project/all_videos_clipped/'):
    '''
    clip the video files 
    '''
    files = os.listdir(path)
    
    for file in files:
        if '.avi' in file:
            target_filename = file
            target_name = os.path.join(target_path,target_filename)
        
            ffmpeg_extract_subclip(os.path.join(path,file), 0, 5, targetname=target_name)
        
        
```

```python
clip_files()
```
